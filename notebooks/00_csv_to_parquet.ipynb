```python

import pandas as pd

# Define las rutas de tus archivos
ruta_csv = "../data/raw/trustpilot-reviews-123k.csv"
ruta_parquet = "../data/processed/trustpilot-reviews-123k.parquet"

# 1. Leer el CSV tal cual está, sin parsear fechas ni alterar nada
# Si tu CSV tiene un separador distinto a la coma, deberás añadir el parámetro sep=';' por ejemplo.
df = pd.read_csv(ruta_csv)

# 2. Exportar a formato parquet
# Usamos index=False para no añadir una columna extra de índice numérico que no existía en el CSV original.
df.to_parquet(ruta_parquet, engine='pyarrow', index=False)

print(f"Archivo convertido exitosamente. Filas: {df.shape[0]}, Columnas: {df.shape[1]}")

#Ahora comparo los dos archivos para asegurarme de que la conversión fue correcta.
# 3. Leer el archivo Parquet
df_parquet = pd.read_parquet(ruta_parquet, engine='pyarrow')

# 4. Comparar los DataFrames
if df.equals(df_parquet):
    print("Los archivos son idénticos.")
else:
    print("Los archivos son diferentes.")

#Imprimo por pantalla algunas filas de ambos DataFrames para inspección visual
print("\nPrimeras filas del DataFrame original (CSV):")
print(df.head())
print("\nPrimeras filas del DataFrame convertido (Parquet):")   
print(df_parquet.head())
```
